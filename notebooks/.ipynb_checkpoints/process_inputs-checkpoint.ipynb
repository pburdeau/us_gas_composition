```python
# USGS

# download geographic data: shale plays, counties, states, basins
us_states = gpd.read_file(shapefiles_path + '/cb_2018_us_state_20m/cb_2018_us_state_20m.shp')

# clean us_counties to keep only the ones in usgs
us_counties = gpd.read_file(shapefiles_path + '/cb_2018_us_county_500k/cb_2018_us_county_500k.shp')
basins_gdf = gpd.read_file(shapefiles_path + '/basins_shapefiles')



index_to_deletes = []
for i in range(len(us_counties)):
    county_name = us_counties['NAME'].iloc[i].upper()
    state_fp = us_counties['STATEFP'].iloc[i]
    if len(us_states[us_states.STATEFP == state_fp]) > 0:
        state_name = us_states[us_states.STATEFP == state_fp].reset_index(drop=True).NAME.iloc[0].upper()
        combined_name = f"{county_name} ({state_name})"
        us_counties['NAME'].iloc[i] = combined_name
        us_counties['COUNTY_NAME'] = county_name
        us_counties['STATE_NAME'] = state_name
    else:
        index_to_deletes.append(i)
        
for i in index_to_deletes:
    us_counties = us_counties.drop(i)


usgs = pd.read_csv(usgs_path)
usgs_not_na = usgs[~pd.isna(usgs.LAT)]
usgs_not_na = usgs_not_na[~pd.isna(usgs.LONG)]
usgs_not_na = usgs_not_na[~pd.isna(usgs.FINAL_SAMPLING_DATE)]

years = []
for i in range(len(usgs_not_na)):
    years.append(int(usgs_not_na.FINAL_SAMPLING_DATE.iloc[i][-4:]))
usgs_not_na['Year'] = years

usgs_gdf = gpd.GeoDataFrame(usgs_not_na, geometry=gpd.points_from_xy(usgs_not_na.LONG, usgs_not_na.LAT), crs="EPSG:4326")

usgs_gdf = usgs_gdf.to_crs(26914)

usgs_gdf = usgs_gdf.rename(columns={'API':'API14'})

usgs_gdf['COUNTY'] = [str(usgs_gdf.COUNTY.iloc[i]) + ' (' + usgs_gdf.STATE.iloc[i] + ')' for i in range(len(usgs_gdf))]

# transform dates in datetime format
dates = []
with tqdm(total=len(usgs_gdf)) as pbar:
    for i in range(len(usgs_gdf)):
        month, day, year = usgs_gdf['FINAL_SAMPLING_DATE'].iloc[i].split('/')
        month = int(month)
        day = int(day)
        year = int(year)
        if month == 0:
            month = 1
        if day == 0:
            day = 1
        if (month == 2) and (day > 28):
            day = 28
        if (day > 30):
            day = 30
        if month > 12:
            month = 12
        dates.append(datetime(year, month, day, 0, 0, 0))
        pbar.update(1)
    usgs_gdf['date'] = dates    

# transforma datetime in number of days since earliest date
usgs_gdf = usgs_gdf.sort_values(by='date').reset_index(drop=True)
epochs = [0]
for i in range(1, len(usgs_gdf)):
    epochs.append((usgs_gdf.date.iloc[i] - usgs_gdf.date.iloc[0]).days)
usgs_gdf['epochs'] = epochs

usgs_gdf = usgs_gdf.drop(columns=['ID', 'SOURCE','BLM ID','USGS ID','ALASKA_QUAD_NAMES','WELL NAME','PUBLICATION DATE','COUNTY','STATE'])

usgs_gdf = gpd.sjoin(usgs_gdf.to_crs(26914), basins_gdf.to_crs(26914), op='within')

usgs_gdf = usgs_gdf.to_crs(26914)

usgs_gdf['X'] = usgs_gdf.geometry.x
usgs_gdf['Y'] = usgs_gdf.geometry.y
usgs_gdf['T'] = np.array(usgs_gdf.epochs)

usgs_gdf = usgs_gdf.drop(columns=['FINAL_DRILLING_DATE', 'FINAL_SAMPLING_DATE', 'SAMPLE DATE BEFORE COMPLETION', 'index_right', 'BASIN_CODE', 'DEPTH','AGE','FORMATION','geometry','FIELD','LAT','LONG'])

numeric_columns = ['HE', 'CO2', 'H2', 'N2', 'H2S', 'AR', 'O2', 'C1', 'C2', 'C3', 'N-C4', 'I-C4', 'N-C5', 'I-C5', 'C6+']

usgs_gdf[numeric_columns] = usgs_gdf[numeric_columns].apply(pd.to_numeric, errors='coerce')
```

```python
def upload_result(all_prod_indices_all, path):
    csv_files_pattern = os.path.join(root_path,
                                     'out',
                                     path)
    csv_files = glob.glob(csv_files_pattern)
    result_all_list = []
    for file in csv_files:
        result = pd.read_csv(file)
        result_all_list.append(result)
    result_all = pd.concat(result_all_list)
    result_all['X'] = result_all['X'] * 6.34
    result_all['Y'] = result_all['Y'] * 6.34
    result_all = result_all.drop(columns='Unnamed: 0')
    merged = pd.merge(all_prod_indices_all, result_all,
                      left_on=['Prod_X_grid', 'Prod_Y_grid', 'Prod_T_grid', 'BASIN_NAME'],
                      right_on=['X', 'Y', 'T', 'BASIN_NAME'])
    columns_to_drop = {'X_y', 'Y_y', 'T_y', 'Prod_index', 'Gas', 'Oil'}
    merged = merged.drop(columns=columns_to_drop.intersection(merged.columns))
    merged = merged.rename(columns={'X_x': 'X_prod', 'Y_x': 'Y_prod', 'T_x': 'T_prod'})
    return merged
```

```python
import seaborn as sns
from scipy.stats import linregress

# Load dataset
df = usgs_prod.copy()

# Compute Gas-to-Oil Ratio (GOR) while handling NaN values
df["GOR"] = df["Monthly Gas"] / df["Monthly Oil"]
df["Log GOR"] = np.log(df["GOR"])

# Handle NaN and infinite values
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.dropna(subset=["Log GOR"], inplace=True)
lower, upper = np.percentile(df["Log GOR"].dropna(), [0, 100])
df_filtered = df[(df["Log GOR"] >= lower) & (df["Log GOR"] <= upper)].copy()

# Define all components
components = ['C1', 'C2', 'C3', 'N-C4', 'I-C4', 'N-C5', 'I-C5', 'C6+']

# Define a color palette
palette = sns.color_palette("crest", n_colors=len(components))[::-1]  # Shades of blue

# Set up subplots without shared axes
n_rows, n_cols = 2, 4  # Adjust grid to fit all components
fig, axes = plt.subplots(n_rows, n_cols, figsize=(12, 6))
axes = axes.flatten()  # Flatten axes for easy iteration

for i, (comp, color) in enumerate(zip(components, palette)):
    ax = axes[i]
    valid_mask = df_filtered["Log GOR"].notna() & df_filtered[comp].notna()
    
    if valid_mask.sum() > 1:
        ax.scatter(df_filtered.loc[valid_mask, "Log GOR"], 
                   df_filtered.loc[valid_mask, comp], 
                   alpha=0.3, color=color, edgecolors="none")

        slope, intercept, r_value, _, _ = linregress(df_filtered.loc[valid_mask, "Log GOR"], df_filtered.loc[valid_mask, comp])
        x_vals = np.linspace(df_filtered["Log GOR"].min(), df_filtered["Log GOR"].max(), 100)
        ax.plot(x_vals, slope * x_vals + intercept, color="black")
        ax.text(0.05, 0.9, f"r={r_value:.3f}", transform=ax.transAxes, bbox=dict(facecolor="white", alpha=0.5))
    
    ax.set_title(comp, fontsize=10, fontweight="bold")
    ax.set_xlabel("Log GOR")
    ax.set_ylabel(f"% {comp}")

plt.tight_layout()

plt.savefig('figures_out/correlation_plots.png', dpi=300)
# Adjust layout to prevent overlap
plt.show()
```

```python
# Define all components
components = ['HE', 'CO2', 'H2', 'N2', 'H2S', 'AR', 'O2','C1', 'C2', 'C3', 'N-C4', 'I-C4', 'N-C5', 'I-C5', 'C6+']
for comp in components:
    valid_mask = df_filtered["Log GOR"].notna() & df_filtered[comp].notna()
    slope, intercept, r_value, _, _ = linregress(df_filtered.loc[valid_mask, "Log GOR"], df_filtered.loc[valid_mask, comp])
    print(f'Comp: {comp}')
    print(f"r={r_value:.3f}")
```

```python
# Define the folder containing CSV files and the new directory for output
input_folder = root_path + '/wells_prod_jan_26'
output_folder = root_path + '/wells_info_prod_per_basin'

# Ensure the output folder exists
os.makedirs(output_folder, exist_ok=True)

# List existing files in the output folder
existing_files = set(os.listdir(output_folder))

# Iterate over all CSV files in the input folder
with tqdm(total=len(os.listdir(input_folder))) as pbar:
    for file in os.listdir(input_folder):
        if file.endswith('.csv') and 'Anadarko' in file:
            print(file)
            # Check if the file already exists in the output folder
#             if file in existing_files:
#                 print(f"File '{file}' already processed. Skipping...")
#                 pbar.update(1)
#                 continue  # Skip this file

            file_path = os.path.join(input_folder, file)

            # Read the CSV file
            df = pd.read_csv(file_path)

            # Add geometry and convert to UTM Zone 14N
            df['geometry'] = df.apply(
                lambda row: Point(row['Surface Hole Longitude (WGS84)'], row['Surface Hole Latitude (WGS84)']),
                axis=1
            )
            gdf = gpd.GeoDataFrame(df, geometry='geometry', crs="EPSG:4326")
            del df
            gdf = gdf.to_crs(26914)  # Convert to UTM Zone 14N
            gdf['X'] = gdf.geometry.x
            gdf['Y'] = gdf.geometry.y
            gdf = gdf.drop(columns='geometry')

            # Convert 'Monthly Production Date' to datetime and extract the year
            gdf['Monthly Production Date'] = pd.to_datetime(gdf['Monthly Production Date'])
            gdf['Year'] = gdf['Monthly Production Date'].dt.year

            # Group by API14, X, Y, and Year, aggregating Monthly Gas and Monthly Oil
            grouped = gdf.groupby(['API14', 'X', 'Y', 'Year'], as_index=False).agg({
                'Monthly Gas': 'sum',
                'Monthly Oil': 'sum',
                'BASIN_NAME': 'first'  # Retain the first value of BASIN_NAME
            })

            # Compute the middle of the year for each group
            grouped['Last Date'] = grouped['Year'].apply(lambda year: pd.Timestamp(f'{year}-12-31'))  # End of the year

            # Calculate T based on the middle date
            grouped['T'] = (grouped['Last Date'] - pd.Timestamp(reference_date)).dt.days

            # Drop the temporary 'Middle Date' column
            grouped = grouped.drop(columns=['Last Date'])

            # Save the grouped data to the output directory
            output_file_path = os.path.join(output_folder, file)
            grouped.to_csv(output_file_path, index=False)
            print(f"Saved grouped data to: {output_file_path}")

        pbar.update(1)
```

```python
def create_grid(df, res_space, res_time):
    xmin, xmax = df['X'].min(), df['X'].max()
    ymin, ymax = df['Y'].min(), df['Y'].max()
    tmin, tmax = df['T'].min(), df['T'].max()
    grid_params = (xmin, xmax, ymin, ymax, tmin, tmax, res_space, res_time)

    # Create 3D grid coordinates
    x_range = np.arange(xmin, xmax + res_space, res_space)
    y_range = np.arange(ymin, ymax + res_space, res_space)
    t_range = np.arange(tmin, tmax + res_time, res_time)
    grid_coord = np.meshgrid(x_range, y_range, t_range, indexing='ij')
    grid = np.column_stack([g.ravel() for g in grid_coord])
    return grid, grid_params

def aggregate(df, list_zz, grid_params, mean_agg=True):
    xmin, xmax, ymin, ymax, tmin, tmax, res_space, res_time = grid_params
    x_range = np.arange(xmin, xmax + res_space, res_space)
    y_range = np.arange(ymin, ymax + res_space, res_space)
    t_range = np.arange(tmin, tmax + res_time, res_time)
    grid_coord = np.meshgrid(x_range, y_range, t_range, indexing='ij')

    df = df[(df.X >= xmin) & (df.X <= xmax)
            & (df.Y >= ymin) & (df.Y <= ymax)
            & (df['T'] >= tmin) & (df['T'] <= tmax)].reset_index(drop=True)
    # Flatten the grid for DataFrame representation
    flat_grid_coords = np.column_stack([g.ravel() for g in grid_coord])

    df_grid = pd.DataFrame(flat_grid_coords, columns=['X', 'Y', 'T'])

    for zz in list_zz:
        list_cols = ['X', 'Y', 'T']
        list_cols.append(zz)

        # Convert DataFrame to NumPy array for processing
        np_data = df[list_cols].to_numpy().astype(float)
        origin = np.array([xmin, ymin, tmin])
        resolution = np.array([res_space, res_space, res_time])

        # Calculate grid indices for each data point
        grid_indices = np.rint(((np_data[:, :3] - origin) / resolution)).astype(int)

        # Initialize 3D aggregation arrays
        grid_shape = (len(x_range), len(y_range), len(t_range))
        grid_sum = np.zeros(grid_shape)
        grid_count = np.zeros(grid_shape)
        # Aggregate data into the grid
        for i in range(np_data.shape[0]):
            x_idx, y_idx, t_idx = grid_indices[i, :3]
            if x_idx < grid_shape[0] and y_idx < grid_shape[1] and t_idx < grid_shape[2]:
                grid_sum[x_idx, y_idx, t_idx] += np_data[i, 3]
                grid_count[x_idx, y_idx, t_idx] += 1

        if mean_agg == True:
            # Calculate average values where count is not zero
            with np.errstate(invalid='ignore', divide='ignore'):
                grid_matrix = np.divide(grid_sum, grid_count)
        #             grid_matrix = np.nan_to_num(grid_matrix)  # Replace NaNs with 0
        else:
            grid_matrix = grid_sum

        flat_grid_matrix = grid_matrix.ravel()

        # update DataFrame
        df_grid[zz] = flat_grid_matrix
    return df_grid


alpha = 5.7
res_space = 2000 / alpha # 1km
res_time = 1825 # 1 years
```

```python
def create_grid_from_df(grid_df, x_unique, y_unique, x_col='X', y_col='Y', value_col='value_column', res_space=None):
    """
    Transforms X, Y, and value_column into a 2D grid (array) format.

    Parameters:
    - grid_df: DataFrame containing X, Y, and value_column columns.
    - x_col: Name of the column representing X coordinates.
    - y_col: Name of the column representing Y coordinates.
    - value_col: Name of the column with values to populate the grid.

    Returns:
    - X_grid: 2D array of X coordinates.
    - Y_grid: 2D array of Y coordinates.
    - Z_grid: 2D array of values from value_column.
    """
    # Create empty 2D array for Z values
    Z_grid = np.full((len(y_unique), len(x_unique)), np.nan)  # Initialize with NaNs

    # Create X and Y coordinate grids using meshgrid
    X_grid, Y_grid = np.meshgrid(x_unique, y_unique)

    # Map each (X, Y) pair to its corresponding value in Z_grid
    for _, row in grid_df.iterrows():
#         x_idx = np.where(x_unique == row[x_col])[0][0]
#         y_idx = np.where(y_unique == row[y_col])[0][0]
        EPS = 1e0
        x_idx = np.argwhere(np.abs(x_unique - res_space / 2. - row[x_col]) < EPS)
        y_idx = np.argwhere(np.abs(y_unique - res_space / 2 - row[y_col]) < EPS)
        Z_grid[y_idx, x_idx] = row[value_col]
    return X_grid, Y_grid, Z_grid

def plot_2d_colormap_pcolormesh(gdf, basins_gdf, value_column, basin_name, res_space, title, cmap, year, source):
    if 'raw' in source:
        linewidth = 10
    else:
        linewidth = 5
    if not isinstance(gdf, gpd.GeoDataFrame):
        if 'X' in gdf.columns and 'Y' in gdf.columns:
            gdf = gpd.GeoDataFrame(gdf, geometry=gpd.points_from_xy(gdf['X'], gdf['Y']))
            gdf.set_crs(basins_gdf.crs, inplace=True)
        else:
            raise ValueError("DataFrame must have 'X' and 'Y' columns to convert to GeoDataFrame.")

    if gdf.crs != basins_gdf.crs:
        gdf = gdf.to_crs(basins_gdf.crs)

    selected_basin = basins_gdf[basins_gdf['BASIN_NAME'] == basin_name][['geometry']]
    gdf = gpd.sjoin(gdf, selected_basin, predicate='within')

    if basin_name == 'all':
        gdf_filtered = gdf
        basin_bounds = basins_gdf.dissolve().geometry.bounds
    else:
        gdf_filtered = gdf[gdf['BASIN_NAME'] == basin_name]
        basin_bounds = basins_gdf[basins_gdf['BASIN_NAME'] == basin_name].geometry.bounds

    x_min_basin, x_max_basin = basin_bounds.minx.values[0], basin_bounds.maxx.values[0]
    y_min_basin, y_max_basin = basin_bounds.miny.values[0], basin_bounds.maxy.values[0]
    extent = [x_min_basin, x_max_basin, y_min_basin, y_max_basin]
    

    # Discretize coordinates for grid with increased resolution (optional)
    gdf_filtered['X_bin'] = np.floor((gdf_filtered['X'] - x_min_basin) / (res_space)) * (res_space) + x_min_basin
    gdf_filtered['Y_bin'] = np.floor((gdf_filtered['Y'] - y_min_basin) / (res_space)) * (res_space) + y_min_basin

    # Aggregate values by averaging within each grid cell
    grid_df = gdf_filtered.groupby(['X_bin', 'Y_bin'])[value_column].mean().reset_index()
    grid_df = grid_df.rename(columns={'X_bin': 'X', 'Y_bin': 'Y'})
    x_unique = np.arange(grid_df.X.min(), grid_df.X.max() + res_space / 2., res_space) + res_space / 2.
    y_unique = np.arange(grid_df.Y.min(), grid_df.Y.max() + res_space / 2., res_space) + res_space / 2.
    
    # Create grids for X, Y, and Z values without using pivot
    X_grid, Y_grid, Z_grid = create_grid_from_df(grid_df, x_unique, y_unique, x_col='X', y_col='Y', value_col=value_column, res_space=res_space)
        
    #if not 'std' in value_column and source != 'prod':
        #min_value_ghgrp = np.nanpercentile(plot_df_ghgrp_prod[plot_df_ghgrp_prod.BASIN_NAME == basin_name][value_column], 5)
        #max_value_ghgrp = np.nanpercentile(plot_df_ghgrp_prod[plot_df_ghgrp_prod.BASIN_NAME == basin_name][value_column], 95)
        #min_value_usgs = np.nanpercentile(plot_df_usgs[plot_df_usgs.BASIN_NAME == basin_name][value_column], 5)
        #max_value_usgs = np.nanpercentile(plot_df_usgs[plot_df_usgs.BASIN_NAME == basin_name][value_column], 95) 
        #min_value = min(min_value_ghgrp, min_value_usgs)
        #max_value = max(max_value_ghgrp, max_value_usgs)
       # min_value = 0.58
       # max_value = 0.96
    #else:
        #min_value = np.nanpercentile(gdf[value_column], 5)
        #max_value = np.nanpercentile(gdf[value_column], 95) 
    min_value = np.nanpercentile(gdf[value_column], 5)
    max_value = np.nanpercentile(gdf[value_column], 95) 
    
    # Plot
    fig, ax = plt.subplots(figsize=(15, 15))
    
    Z_grid[np.isnan(Z_grid)] = np.inf

    # Use pcolormesh to plot the colormap with `nearest` shading
    c = ax.pcolor(X_grid, Y_grid, Z_grid, cmap=cmap, shading='auto', snap=True, vmin=min_value, vmax=max_value, edgecolors=None, linewidth=-1)
    
    
    # Plot the basin boundary on top to cover any edge artifacts
    basins_gdf[basins_gdf['BASIN_NAME'] == basin_name].plot(ax=ax, color='none', edgecolor='black',
                                                            linestyle='dashed', linewidth=linewidth)

    # Skip colorbar here if 'std' in value_column
    if 'std' in value_column:
        # Save the colormap as a standalone plot
        fig_colorbar, ax_colorbar = plt.subplots(figsize=(2, 10))
        cbar = fig_colorbar.colorbar(c, ax=ax_colorbar, orientation='vertical', shrink=0.6, aspect=15)
        cbar.set_label(f'{value_column} [mol %]', fontsize=20, labelpad=10, loc='top')
        cbar.ax.tick_params(labelsize=20)
        ax_colorbar.set_frame_on(False)  
        ax_colorbar.axis("off")         

        fig_colorbar.savefig(f"figures_out/colorbar_{value_column}_{source}_{basin_name}_{year}_{res_space}.eps",
                             format='eps', dpi=300, bbox_inches='tight')
        plt.close(fig_colorbar)

    
    # Set axis limits based on the basin bounds
    ax.set_xlim(x_min_basin, x_max_basin)
    ax.set_ylim(y_min_basin, y_max_basin)

    # Remove grid lines, ticks, and spines
    ax.grid(False)
    ax.set_xticks([])
    ax.set_yticks([])
    for spine in ax.spines.values():
        spine.set_visible(False)
    plt.savefig(f"figures_out/figure_{value_column}_{source}_{basin_name}_{year}_{res_space}.eps", format='eps', dpi=300, bbox_inches='tight')
    plt.show()
    
def plot_2d_colormap_pcolormesh_with_legend(gdf, basins_gdf, value_column, basin_name, res_space, title, cmap, year, source):
#     if 'raw' in source:
#         linewidth = 10
#     else:
    linewidth = 5
    if not isinstance(gdf, gpd.GeoDataFrame):
        if 'X' in gdf.columns and 'Y' in gdf.columns:
            gdf = gpd.GeoDataFrame(gdf, geometry=gpd.points_from_xy(gdf['X'], gdf['Y']))
            gdf.set_crs(basins_gdf.crs, inplace=True)
        else:
            raise ValueError("DataFrame must have 'X' and 'Y' columns to convert to GeoDataFrame.")

    if gdf.crs != basins_gdf.crs:
        gdf = gdf.to_crs(basins_gdf.crs)

    selected_basin = basins_gdf[basins_gdf['BASIN_NAME'] == basin_name][['geometry']]
    gdf = gpd.sjoin(gdf, selected_basin, predicate='within')

    if basin_name == 'all':
        gdf_filtered = gdf
        basin_bounds = basins_gdf.dissolve().geometry.bounds
    else:
        gdf_filtered = gdf[gdf['BASIN_NAME'] == basin_name]
        basin_bounds = basins_gdf[basins_gdf['BASIN_NAME'] == basin_name].geometry.bounds

    x_min_basin, x_max_basin = basin_bounds.minx.values[0], basin_bounds.maxx.values[0]
    y_min_basin, y_max_basin = basin_bounds.miny.values[0], basin_bounds.maxy.values[0]
    extent = [x_min_basin, x_max_basin, y_min_basin, y_max_basin]

    # Discretize coordinates for grid with increased resolution (optional)
    gdf_filtered['X_bin'] = np.floor((gdf_filtered['X'] - x_min_basin) / res_space) * res_space + x_min_basin
    gdf_filtered['Y_bin'] = np.floor((gdf_filtered['Y'] - y_min_basin) / res_space) * res_space + y_min_basin

    # Aggregate values by averaging within each grid cell
    grid_df = gdf_filtered.groupby(['X_bin', 'Y_bin'])[value_column].mean().reset_index()
    grid_df = grid_df.rename(columns={'X_bin': 'X', 'Y_bin': 'Y'})
    x_unique = np.arange(grid_df.X.min(), grid_df.X.max() + res_space / 2., res_space) + res_space / 2.
    y_unique = np.arange(grid_df.Y.min(), grid_df.Y.max() + res_space / 2., res_space) + res_space / 2.

    # Create grids for X, Y, and Z values without using pivot
    X_grid, Y_grid, Z_grid = create_grid_from_df(grid_df, x_unique, y_unique, x_col='X', y_col='Y', value_col=value_column, res_space=res_space)

    min_value = np.nanpercentile(gdf[value_column], 5)
    max_value = np.nanpercentile(gdf[value_column], 95)

    # Plot
    fig, ax = plt.subplots(figsize=(15, 15))

    Z_grid[np.isnan(Z_grid)] = np.inf

    # Use pcolormesh to plot the colormap with `nearest` shading
    c = ax.pcolormesh(X_grid, Y_grid, Z_grid, cmap=cmap, shading='auto', snap=True, vmin=min_value, vmax=max_value)

    # Plot the basin boundary on top to cover any edge artifacts
    basins_gdf[basins_gdf['BASIN_NAME'] == basin_name].plot(ax=ax, color='none', edgecolor='black',
                                                            linestyle='dashed', linewidth=linewidth)

    # Add the colorbar within the same figure
    cbar = fig.colorbar(c, ax=ax, orientation='vertical', shrink=0.6, aspect=15, pad=0.02)
    cbar.set_label(f'{value_column} [mol %]', fontsize=20, labelpad=10, loc='top')
    cbar.ax.tick_params(labelsize=20)

    # Set axis limits based on the basin bounds
    ax.set_xlim(x_min_basin, x_max_basin)
    ax.set_ylim(y_min_basin, y_max_basin)

    # Remove grid lines, ticks, and spines
    ax.grid(False)
    ax.set_xticks([])
    ax.set_yticks([])
    for spine in ax.spines.values():
        spine.set_visible(False)

    plt.savefig(f"figures_out/figure_{value_column}_{source}_{basin_name}_{year}_{res_space}.eps", format='eps', dpi=300, bbox_inches='tight')
    #plt.show()
```

```python
alpha = 5.7
res_space = 2000 / alpha # 1km
res_time = 1825 # 1 years
n_closest_years = 10

filtered_dfs_prod = []
basin_names = ['Anadarko Basin']
with tqdm(total=len(basin_names)) as pbar:
    for basin_name in basin_names:
        wells_data = grouped  # only basin data
        df_prod = wells_data.reset_index(drop=True)
        df_prod = df_prod.rename(columns={'Monthly Oil': 'Oil', 'Monthly Gas': 'Gas'})
        df_prod = df_prod[df_prod.Gas > 0].reset_index(drop=True)
        df_prod['X'] = df_prod.X / alpha
        df_prod['Y'] = df_prod.Y / alpha
        df_prod = df_prod[~pd.isna(df_prod['Gas'])].reset_index(drop=True)

        grid, grid_params = create_grid(df_prod, res_space, res_time)

        prod_on_grid = aggregate(df_prod, ['Gas'], grid_params, mean_agg=False)

        reference_date = '1918-08-19T00:00:00.000000000'
        prod_on_grid['Year'] = (pd.Timestamp(reference_date) + pd.to_timedelta(data_on_grid['T'], unit='D')).dt.year
        
        year = 2017
        prod_on_grid['BASIN_NAME'] = basin_name
        basin_df_prod = prod_on_grid.copy()

        # Ensure only unique years are considered
        unique_years_prod = basin_df_prod['Year'].drop_duplicates().sort_values()

        sorted_indices_prod = (unique_years_prod - year).abs().sort_values().index[:n_closest_years]

        # Select the closest years
        closest_years_prod = unique_years_prod.loc[sorted_indices_prod].values

        basin_filtered_df_prod = basin_df_prod[basin_df_prod['Year'].isin(closest_years_prod)]

        filtered_dfs_prod.append(basin_filtered_df_prod)
        pbar.update(1)

plot_df_prod_problem_original_data  = pd.concat(filtered_dfs_prod, ignore_index=True)

plot_df_prod_problem_original_data['X'] = plot_df_prod_problem_original_data['X'] * 5.7
plot_df_prod_problem_original_data['Y'] = plot_df_prod_problem_original_data['Y'] * 5.7
```
